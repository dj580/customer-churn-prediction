```python
# converting the 'TotalCharges' column to numeric, coercing errors to NaN

teleco_cust['TotalCharges'] = teleco_cust['TotalCharges'].replace(' ', np.nan)
```

```python
# converting the 'TotalCharges' column to numeric, coercing errors to NaN
teleco_cust['TotalCharges'] = pd.to_numeric(teleco_cust['TotalCharges'], errors='coerce')   
```

```python
#Removing missing values 

teleco_cust.dropna(inplace = True)
```

```python
teleco_cust.isna().sum()
```

```python
#Remove customer IDs from the data set

df = teleco_cust.iloc[:,1:]
```

```python
df.head()
```

```python
#Convertin the predictor variable in a binary numeric variable

df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
df.head()

```

```python
#Splitting the data into features and target variable
x = df.drop("Churn", axis=1)
y = df["Churn"]

```

```python
#Splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
#Encoding categorical variables using one-hot encoding
x_train = pd.get_dummies(x_train, drop_first=True) #one-hot encoding for training data
x_test  = pd.get_dummies(x_test, drop_first=True)  #one-hot encoding for testing data

```

```python
# ALIGNMENT COMPLETE
#Ensure both train and test sets have the same columns after one-hot encoding
x_train, x_test = x_train.align(x_test, join='left', axis=1, fill_value=0)
```

```python
x_train
```

```python
x_test
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

num_cols = ["tenure", "MonthlyCharges", "TotalCharges"]

scaler = StandardScaler()

# Fit and transform training data
x_train[num_cols] = scaler.fit_transform(x_train[num_cols])

# Transform testing data
x_test[num_cols] = scaler.transform(x_test[num_cols])
```

Task 2: Model Building 

```python
#Training a Logistic Regression model
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train) # fitting the model

y_pred_lr = lr.predict(x_test)
```

```python
#Training a Random Forest Classifier model
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=1000,
    random_state=42
)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)
```

```python
#Evaluating the Logistic Regression model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Logistic Regression Evaluation")
print("Accuracy :", accuracy_score(y_test, y_pred_lr))
print("Precision:", precision_score(y_test, y_pred_lr))
print("Recall   :", recall_score(y_test, y_pred_lr))
print("F1 Score :", f1_score(y_test, y_pred_lr))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Get feature names and coefficients
feature_importance = pd.DataFrame({
    "Feature": x_train.columns,
    "Coefficient": lr.coef_[0]
})

# Calculate absolute importance
feature_importance["Absolute_Importance"] = feature_importance["Coefficient"].abs()

# Sort by importance
feature_importance = feature_importance.sort_values(
    by="Absolute_Importance",
    ascending=False
)

# Display top 10 important features
print(feature_importance.head(10))

# Plot top 10 feature importance
top_features = feature_importance.head(10)

plt.figure(figsize=(8, 5))
plt.barh(top_features["Feature"], top_features["Coefficient"])
plt.xlabel("Coefficient Value")
plt.title("Top 10 Feature Importance (Logistic Regression)")
plt.gca().invert_yaxis()
plt.show()
```

```python
#Evaluating the Random Forest Classifier model
print("\nRandom Forest Evaluation")
print("Accuracy :", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf))
print("Recall   :", recall_score(y_test, y_pred_rf))
print("F1 Score :", f1_score(y_test, y_pred_rf))
```
